```python
# libraries!
import numpy as np      # numpy is Python's "array" library
import pandas as pd     # Pandas is Python's "data" library ("dataframe" == spreadsheet)
```

```python
df = pd.read_csv('titanic.csv')        
df
```

```python
df.columns
```

```python
ROW = 0
COLUMN = 1

df_clean1 = df.drop('the original Kaggle dataset!   Adapted from here:   https://www.kaggle.com/c/titanic', axis=COLUMN)
df_clean1
```

```python
# Drop columns with words

df_clean2 = df_clean1.drop(['name', 'ticket', 'fare', 'cabin', 'embarked', 'boat', 'body', 'home.dest'], axis=COLUMN)
df_clean2
```

```python
df_clean3 = df_clean2.dropna()
df_clean3
```

```python
SEX = ['female','male']   # int to str
SEX_INDEX = {'female':0,'male':1}  # str to int

def convert_sex(sex):
    """ return the species index (a unique integer/category) """
    #print(f"converting {speciesname}...")
    return SEX_INDEX[sex]

df_clean4 = df_clean3.copy()  # copy everything AND...

df_clean4['sexnum'] = df_clean3['sex'].apply(convert_sex)

# let's see...
df_clean4
```

```python
df_clean4.to_csv('titanic_cleaned.csv', index_label=False)
```
