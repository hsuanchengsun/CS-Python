```python
# libraries!
import numpy as np      # numpy is Python's "array" library
import pandas as pd     # Pandas is Python's "data" library ("dataframe" == spreadsheet)
```

```python
df = pd.read_csv('digits.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
ROW = 0
COLUMN = 1

df_clean1 = df.drop('excerpted from http://yann.lecun.com/exdb/mnist/', axis=COLUMN)
df_clean1
```

```python
df_clean2 = df_clean1.dropna()
df_clean2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

row_to_show = 42
numeral = df_clean2.loc[row_to_show,'actual_digit']     # example of access by row, col with a column name
numeral = int(round(numeral))                    # not really needed
print(f"The numeral itself is a +++  {numeral}  +++\n")

pixels = df_clean2.iloc[row_to_show,0:64]  # dont want the rightmost rows
pixels = pixels.values              # converts to numpy array
pixels = pixels.astype(int)         # convert to integers for plotting
pixels = np.reshape(pixels, (8,8))  # makes a series of 64 values into an 8x8 grid
print(f"Its pixels are\n{pixels}")  

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
# plot! Change annot=True to False to skip the numbers...  

sns.heatmap(pixels, annot=True, fmt="d", linewidths=.5, ax=ax, cmap="Purples")
```

```python
df_tidy =  df_clean2

# Now, save
df_tidy.to_csv('digits_clean.csv', index_label=False)
```
