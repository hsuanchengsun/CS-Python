```python
# libraries!
import numpy as np      # numpy is Python's "array" library
import pandas as pd     # Pandas is Python's "data" library ("dataframe" == spreadsheet)
```

```python
# Read Data

df = pd.read_csv('births.csv')        # encoding="utf-8" et al.
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns
```

```python
# Drop the last column

df_clean1 = df.drop(df.columns[-1], axis=1)
df_clean1.head()
```

```python
# Investigate wrong day

df_clean2 = df_clean1.sort_values(by = ['births']).copy()
```

```python
df_clean2.reset_index(inplace=True)
```

```python
df_clean2.head(10)
```

```python
# Drop 0:5

df_clean3 = df_clean2.drop([0,1,2,3,4,5], axis = 0)
df_clean3
```

```python
# Drop NA

df_clean4 = df_clean3.dropna().copy()
```

```python
# Convert str to num

MEDIAN = ['below','above']   # int to str
MEDIAN_INDEX = {'below':0,'above':1}  # str to int

def convert_median(median):
    return MEDIAN_INDEX[median]

df_clean4['mediannum'] = df_clean4['above/below median'].apply(convert_median)
```

```python
df_clean4
```

```python
# drop the original index

df_clean5 = df_clean4.drop(df_clean4.columns[0], axis=1)
df_clean5
```

```python
# Save file

df_clean5.to_csv('births_cleaned.csv', index_label=False)
```
