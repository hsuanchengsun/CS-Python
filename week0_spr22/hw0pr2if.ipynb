```python
#
# hw0pr2if.py
#

""" 
Title for your adventure:   The Quest.

Notes on how to "win" or "lose" this adventure:
  To win, choose the table.
  To lose, choose the door.
"""

import time

def adventure():
    """This function runs one session of interactive fiction
       Well, it's "fiction," depending on the pill color chosen...
       Arguments: no arguments (prompted text doesn't count as an argument)
       Results: no results     (printing doesn't count as a result)
    """
    delay = 1.0          # change to 0.0 for testing or speed runs,
                         # ..larger for dramatic effect!

    user_name = input("What do they call you, worthy adventurer? ")
    
    while True:
        print()
        print("GOD: Welcome,", user_name, ", to the Death Kindom.")
        print("GOD: Here, you will be tested to see if you have enough wisdom and strength")
        print("to be the brave warrior and save the world.")
        print()

        print("GOD: Your quest: Safely pass all the tests and survive!")
        print()

        print("GOD: Your first mission: choose the correct door!")
        print("GOD: The instinct to identify dangers is one of the most important")
        print("skills braver should have in the adventure.")
        print("GOD: Here are two doors, you should choose you would like to open")
        print("the left one or the right one.")
        print("GOD: Use your instinct to avoid dangers and choose the correct door.")
        time.sleep(delay)
        print()

        choice1 = input("The door you gonna open is: ")
        print()

        if choice1 == "left":
            print("You walk to the left door and open it.")
            print("Suddenly")
            time.sleep(delay)
            print("Lots of monsters pop out from the door and attack you.")
            print("They show no mercy to you and this is the end of your journey.")
            break
        elif choice1 == "right":
            print("You walk to the right door and open it.")
            time.sleep(delay)
            print("Suddenly, a wooden bridge pops out in front of you.")
            print()
            print("GOD: Well done, warrior. Go ahead to the next mission.")
        else:  
            print("There is no other exit in the room.")
            print("Make your choice carefully.")
            break
            
        print()
        print("When you walk to the end of the bridge and open the door,")
        print("a zombie rushes to you and try to bite you.")
        print("You have to take an action to dodge the attack immediately!")
        print("You can either jump or get down.")
        print()
        
        choice2 = input("Choose the way you would like to dodge: ")
        print()

        if choice2 == "jump":
            print("You jump high to dodge the zombie.")
            print("The zombie just fall down from the bridge.")
            print()
            print("GOD: Good job, warrior! You are so brave!")
            print("GOD: Now we move to the next mission.")

        elif choice2 == "get down":
            print("You get down to dodge the zombie.")
            print("The zombie just fall down from the bridge.")
            print("However, it grabs your leg and pull you down from the bridge.")
            break

        print()
        print("After beating zombie, you take an elevator down to the ground.")
        print("You find there are lots of trees around you.")
        print("GOD: You are in the center of the Misty Forest.")
        print("GOD: Choose a direction to escape out of the forest.")
        print()

        choice3 = input("Select your direction from east, west, north, or south: ")
        print()

        if choice3 == "west":
            print("You choose the west side to escape out of the forest.")
            print("On the way, you always hear whispers from the dark side in the forest.")
            print("Fortunately, there is nothing happen.")
            print("Finally, you see a bright exit at the end of the forest.")

        elif choice3 == "east":
            print("You choose the east side to escape out of the forest.")
            print("On the way, you always hear whispers from the dark side in the forest.")
            print("Suddenly, several werewolves show up in front of you.")
            print("They start to hunt you and quickly catch you.")
            print("Your journey ends up here.")
            break
        elif choice3 == "north":
            print("You choose the north side to escape out of the forest.")
            print("On the way, you always hear whispers from the dark side in the forest.")
            print("Suddenly, several demons show up in front of you.")
            print("They start to hunt you and quickly catch you.")
            print("Your journey ends up here.")
            break
        elif choice3 == "south":
            print("You choose the south side to escape out of the forest.")
            print("On the way, you always hear whispers from the dark side in the forest.")
            print("Suddenly, several evil spirits show up in front of you.")
            print("They start to hunt you and quickly catch you.") 
            print("Your journey ends up here.")
            break
        else:  
            print("You stay in the original place and do not move.")
            print("After a while, you find trees get closer to you and there is no way out.")
            print("Finally, you are swallowed by the forest.")
            break

        print()
        print("GOD: You have successfully passed all the missions.")
        print("GOD: You are a brave warrior now.")
        print("GOD: Please choose your reward from these two treasure boxes.")
        print("GOD: Either the big one or the small one.")
        print()
        
        choice4 = input("Which box you would like to select: ")
        print()

        if choice4 == "small":
            print("GOD: You are a humble person.")
            print("GOD: I put the best armor and sword in the box.")
            print("GOD: They could support you while saving the world.")
            print("GOD: Go ahead, brave warrior.")

        else:  
            print("GOD: You are so greedy.")
            print("GOD: The treasure is not the target for you to save the world.")
            print("GOD: You fail the test, get out of here!")
            print("You are kicked out by the god.")
            break
            
        print()
        choice5 = input("Enter yes to go back to your world.")
        print()

        if choice5 == "yes":
            print("A strong white light explode and you could not see anything...")
            time.sleep(delay)
            print("MOM: Hey!! Wake up!! It is already noon!")
            print("Your mom is yelling on you.")
            print("MOM: You have already slept for more than 12 hours!")
            print("MOM: Wake up and do your homework!")
        break

# This next line _runs_ the adventure function, defined just above.
# It's 100% ok to define-then-run all in one notebook cell.

adventure()
```
