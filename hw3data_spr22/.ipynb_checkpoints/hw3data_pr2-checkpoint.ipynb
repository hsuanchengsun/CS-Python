
#### example of a directly web-copied visualization
+ This is from the seaborn gallery
+ It shows off a "heatmap" visualization
+ and uses the built-in <i>flights</i> data

<b>Warning:</b> This example has <i><u>considerably</u></i> more transformation than you need for your own visualizations.

You will likely need <i><b>some</b></i>, and you should customize <i>something</i>.

However, this example is here to emphasize
+ including the original and your customized version
+ how to visualize the digits (important!)

```python
#
# Making sure a provided example works, as-is...
#

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()  # sns.set_theme()

# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9,6))     # create axes of the desired size
sns.heatmap(flights, annot=False, fmt="d", linewidths=.5, ax=ax)

# Interesting... perhaps change the annot=False to annot=True...
```

```python
# Now, let's adapt the above to use our digits data...
# 
# for read_csv, use header=0 when row 0 is a header row
# 
filename = 'digits.csv'
digits_orig = pd.read_csv(filename)   
print(f"{filename} : file read into a pandas dataframe.")
digits_orig
```

```python
# That last column isn't really data at all...
COLUMN = 1; ROW = 0
digits = digits_orig.drop( 'excerpted from http://yann.lecun.com/exdb/mnist/', axis=COLUMN )
digits
```

```python
#
# Let's grab one row and shape it to be 8x8 (which is shape (8,8))
#
googlesheet_row_to_show = 900    # remember this will be different in the spreadsheet

row_to_show = googlesheet_row_to_show - 2

# numeral = df.iloc[row_to_show,64])             # example of using the column from integer index
numeral = digits.loc[row_to_show,'actual_digit']     # example of access by row, col with a column name
numeral = int(round(numeral))                    # not really needed
print(f"The numeral itself is a +++  {numeral}  +++\n")

pixels = digits.iloc[row_to_show,0:64]  # dont want the rightmost rows
pixels = pixels.values              # converts to numpy array
pixels = pixels.astype(int)         # convert to integers for plotting
pixels = np.reshape(pixels, (8,8))  # makes a series of 64 values into an 8x8 grid
print(f"Its pixels are\n{pixels}")  

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
# plot! Change annot=True to False to skip the numbers...  

sns.heatmap(pixels, annot=True, fmt="d", linewidths=.5, ax=ax, cmap="Purples")
# all seaborn palettes: medium.com/@morganjonesartist/color-guide-to-seaborn-palettes-da849406d44f
# cmap=None for reddish palette, cmap="Accent", cmap="gray_r" (best?), cmap="Purples", cmap="spam" to see the list...
```

```python
# Original

sns.set_theme()

# Load the brain networks example dataset
df = sns.load_dataset("brain_networks", header=[0, 1, 2], index_col=0)

# Select a subset of the networks
used_networks = [1, 5, 6, 7, 8, 12, 13, 17]
used_columns = (df.columns.get_level_values("network")
                          .astype(int)
                          .isin(used_networks))
df = df.loc[:, used_columns]

# Create a categorical palette to identify the networks
network_pal = sns.husl_palette(8, s=.45)
network_lut = dict(zip(map(str, used_networks), network_pal))

# Convert the palette to vectors that will be drawn on the side of the matrix
networks = df.columns.get_level_values("network")
network_colors = pd.Series(networks, index=df.columns).map(network_lut)

# Draw the full plot
g = sns.clustermap(df.corr(), center=0, cmap="vlag",
                   row_colors=network_colors, col_colors=network_colors,
                   dendrogram_ratio=(.1, .2),
                   cbar_pos=(.02, .32, .03, .2),
                   linewidths=.75, figsize=(12, 13))

g.ax_row_dendrogram.remove()
```

```python
# Original

sns.set_theme(style="whitegrid")

# Load the example planets dataset
planets = sns.load_dataset("planets")

cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=planets,
    x="distance", y="orbital_period",
    hue="year", size="mass",
    palette=cmap, sizes=(10, 200),
)
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

```python
# Origin

sns.set_theme(style="whitegrid")

diamonds = sns.load_dataset("diamonds")
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]

sns.boxenplot(x="clarity", y="carat",
              color="b", order=clarity_ranking,
              scale="linear", data=diamonds)
```
